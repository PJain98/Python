```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r-') # 'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)           #[ 1*2 means 2 plots]
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(2,2,1)     # [2*2 means total 4 plots]
plt.plot(x, y, 'r--') # More on color options later
plt.subplot(2,2,2)
plt.plot(y, x, 'g*-');
plt.subplot(2,2,3)
plt.plot(y, x, 'g-');
plt.subplot(2,2,4)
plt.plot(y, x, 'g*');
```

## Matplotlib Object Oriented Method

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes (larger plot)
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes (smaller plot)

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

## subplots()

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Axes is an array of axes to plot on
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
fig
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x,y,"g")
axes[1].plot(y,x,"r-")
fig
```

## Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()   #ax.legend(loc=0)   --> loc is not compulsory and 0 means best and we can choose b/w 0 to 10 
                   #and we also can do 
                   # like loc=(0.1,0.1)
```

```python
# Lots of options....

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

# Setting colors, linewidths, linetypes

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

```python

fig = plt.figure()
axes= fig.add_axes([0,0,1,1])
axes.plot(x, x**2, 'r.-') # red line with dots
axes.plot(x, x**3, 'g--') # green dashed line
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.3)              # (alpha is for collour intensity )
ax.plot(x, x+2, color="#8B008B", linewidth=10)        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")                      # RGB hex code 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```

```python
# (we are using same figure and add different line curves)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
fig
```

```python
# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)     
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
fig
```

```python
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
fig
```

```python
# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
fig
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, y, color="red", lw=3, ls='--', marker='o', markersize=15, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Control over axis appearance

```python
fig = plt.figure(figsize=(3,2))
axes= fig.add_axes([0,0,1,1])

axes.plot(x, y, "r--")
axes.set_xlim([3,4])
```

```python
axes.set_xlim([1,4])
fig
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([3, 5])
axes[2].set_title("custom axes range");
```
