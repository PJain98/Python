```python
tips.head()
```

```python
x = tips['total_bill']
y = tips['tip']

sns.jointplot(x = x, y=y)
```

```python
sns.set()
sns.jointplot(x = x, y=y, kind = 'hex')
```

```python
sns.jointplot(x = x, y = y, kind = 'kde')   #kernel density estimate(kde)
```

```python
g = sns.jointplot(x, y, kind = 'kde', color = 'r')
g.plot_joint(plt.scatter, c = 'b', s = 30, linewidth = 1, marker = '+')
g.ax_joint.collections[0].set_alpha(0)
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels = 10)
```
