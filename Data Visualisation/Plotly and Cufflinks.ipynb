```python
df = pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df
```

```python
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```
